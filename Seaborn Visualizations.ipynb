```python
df5=sns.load_dataset('tips')
df5.head()
```

```python
sns.catplot(x = 'day' , y = 'tip' , data =df5 )
```

```python
sns.catplot(x = 'day' , y = 'tip' , kind = 'box' , data = df5 , hue = 'sex')
```

```python
sns.jointplot(x = df5['total_bill'] , y =df5['tip'])
```

```python
sns.jointplot(x = df5['total_bill'] , y =df5['tip'], kind = 'hex')
```

```python
sns.pairplot(df5)
```

```python
sns.lmplot(x = 'total_bill'  , y = 'tip' , data= df5)
```

```python
sns.regplot(x = 'total_bill'  , y = 'tip' , data= df5)
```

```python
sns.distplot(df5['total_bill'],kde=True)
```

```python
sns.distplot(df5['total_bill'],kde=True,hist=False)
```
