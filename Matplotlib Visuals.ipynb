```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
X=[1,3,5,7,9,11,13,15]
Y=[8,16,24,32,40,48,60,70]
plt.plot(X,Y)
plt.title("Simple Linear")
plt.xlabel('X values') 
plt.ylabel('Y values')
plt.grid()
plt.show()
```

```python
plt.bar(X,Y)
plt.xlabel('X values') 
plt.ylabel('Y values')
plt.grid()
plt.show()
```
